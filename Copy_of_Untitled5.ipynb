```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import random

# Sample text from the provided explanation
text = """
Emphasize Rhetorical Awareness in technical communication to teach ethos, pathos, logos.
Contextualize the rhetorical principles within the technical communication genres and settings.
Focus on audience-centered design for clarity, precision, and accessibility in technical documents.
Practice technical writing assignments, incorporating case studies and peer reviews.
Incorporate technology and digital media tools such as web writing, content management systems.
Reflect on the ethical considerations of technical communication like accessibility and transparency.
"""

# Function to generate word cloud
def generate_word_cloud(text):
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(text)
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.show()

# Display the word cloud
generate_word_cloud(text)

# List of True/False questions based on key points
questions = [
    ("Technical communication should emphasize rhetorical awareness, including ethos, pathos, and logos.", True),
    ("In technical communication, only the purpose of the document matters, not the audience.", False),
    ("Technical writing should include hands-on practice and case studies to reinforce learning.", True),
    ("Reflecting on ethical considerations such as transparency and accessibility is not necessary in technical communication.", False),
    ("Technology and digital tools are irrelevant to modern technical communication practices.", False),
    ("Audience-centered design is key for clarity, precision, and accessibility in technical writing.", True)
]

# Function to run the True/False game
def true_false_game(questions):
    score = 0
    random.shuffle(questions)  # Shuffle questions for variety
    for question, correct_answer in questions:
        print(f"Question: {question}")
        user_answer = input("Answer (True/False): ").strip().lower()

        # Check user's answer
        if (user_answer == 'true' and correct_answer) or (user_answer == 'false' and not correct_answer):
            print("Correct!\n")
            score += 1
        else:
            print("Incorrect.\n")

    print(f"Your score: {score}/{len(questions)}")

# Run the True/False game
true_false_game(questions)
```
